```python
%matplotlib inline
import pandas as pd
from pandas.plotting import scatter_matrix
```

## Deleting rows with missing data

```python
raw_data_count = recent_grads.shape
recent_grads = recent_grads.dropna()
```

```python
cleaned_data_count = recent_grads.shape
print(raw_data_count)
print(cleaned_data_count)
```

## Exploring Data in Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter'
                 , title='Median Salary vs Sample size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter'
                 , title='Unemployment Rate vs Sample size')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter'
                 , title='Median Salary vs Full Time')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter'
                 , title='ShareWomen vs Unemployment Rate')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter'
                 , title='Median Salary vs Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter'
                 , title='Median Salary vs Women')
```

## Questions

**1. Do students in more popular majors make more money?**  

    Majors with the largest sample size skew towards lower salaries
    

**2. Do students that majored in subjects that were majority female make more money?**  

    Majors with majority female students tended to earn less money.
    

**3. Is there any link between the number of full-time employees and median salary?**  

    Appears to be a gaussian distribution

## Revisting Questions With Scatter Matrix Plots

**1. Do students in more popular majors make more money?**  

    Majors with the largest sample size skew towards lower salaries while smaller sample sizes correlate with higher salaries.
    

**2. Do students that majored in subjects that were majority female make more money?**  

    Majors with majority female students tended to earn less money.
    

**3. Is there any link between the number of full-time employees and median salary?**
