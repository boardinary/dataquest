## Renaming Columns

The dataset contains mostly string objects and a few inteeger types. Some columns have a small percentage of null values. Column names use camelcase. We will change the column names to snake case by copying the dataframe columns array attribute to a list, edit the elements and assign back to the autos.columns attribute.

```python
autos.columns
```

```python
# converting column names to snake case
new_columns = autos.columns.copy().tolist()
new_columns[0] = "date_crawled"
new_columns[3] = "offer_type"
new_columns[6] = "vehicle_type"
new_columns[7] = "registration_year"
new_columns[9] = "power_PS"
new_columns[12] = "registration_month"
new_columns[13] = "fuel_type"
new_columns[15] = "unrepaired_damage"
new_columns[16] = "ad_created"
new_columns[17] = "nr_of_pictures"
new_columns[18] = "postal_code"
new_columns[19] = "last_seen"
autos.columns = new_columns
autos.head()
```

```python
autos.describe()
```

## Initial Observations:

Columns to drop that are mostly one value:  
- `offer_type` is mostly "Angiebot"
- `seller` is private
- `nr_of_pictures` is entirely 0

Numeric data stored as text:
- all date columns
- odometer
- price

```python
# Converting columns with numeric strings to numeric types
autos['price'] = (autos['price']
                  .str.replace('$', '')
                  .str.replace(',', '')
                  .astype(int))
autos['odometer'] = (autos['odometer']
                     .str.replace('km', '')
                     .str.replace(',', '')
                     .astype(int))
autos.rename({"odometer" : "odometer_km"}, axis=1, inplace=True)
```

## Exploring Mean Price by Brand

```python
autos['brand'].value_counts(normalize=True)
```

We will analyze the brands with greater than a 5% market share. We will loop over these brands to create a dict that contains the brand and mean price.

```python
brands = autos['brand'].value_counts().head(6).index
mean_price_dict = {}

for brand in brands:
    mean_price_dict[brand] = autos.loc[autos['brand'] == brand, 'price'].mean()

print(mean_price_dict)
```

The brands can be broken into three categories:  

Expensive
- Audi
- Mercedes Benz
- BMW

Moderate
- VW

Value
- Opel
- Ford

Of these brands, VW is the best selling.

Next we will use a similar analysis to look at mean mileage for the top brands

```python
brand_mean_mileage = {}
for brand in brands:
    brand_mean_mileage[brand] = autos.loc[autos['brand'] == brand, 'odometer_km'].mean()

print(brand_mean_mileage)
```

We will now compare the two analyses by combining them into a dataframe

```python
bmp_series = pd.Series(mean_price_dict)
bmm_series = pd.Series(brand_mean_mileage)

df = pd.DataFrame(bmp_series, columns=['mean_price'])
df['mean_mileage'] = bmm_series
df
```

Audi's, while the most expensive brand, have the least mileage of the expensive brands. VW, the most popular brand, has the second lowest mileage. Ford, the least popular brand has the lowest mileage.
